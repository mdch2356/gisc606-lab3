```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
import rioxarray as rxr 
```

```python
from rasterio.plot import plotting_extent
```

```python
import earthpy as et
import earthpy.plot as ep
```

```python
sns.set(font_scale=1.5, style="white")
```

```python
et.data.get_data("colorado-flood")
os.chdir(os.path.join(et.io.HOME, 'earth-analytics', 'data'))
```

```python
lidar_dem_path = os.path.join("colorado-flood", 
                              "spatial", 
                              "boulder-leehill-rd", 
                              "pre-flood", 
                              "lidar", 
                              "pre_DTM.tif")

lidar_dem_im = rxr.open_rasterio(lidar_dem_path, masked=True)

lidar_hs_path = os.path.join("colorado-flood", 
                             "spatial", 
                              "boulder-leehill-rd", 
                             "pre-flood", 
                              "lidar", 
                             "pre_DTM_hill.tif")

lidar_dem_hill = rxr.open_rasterio(lidar_hs_path, masked=True)
```

```python
# Challenge 1: Open And Plot Hillshade

fig, ax = plt.subplots(figsize=(10, 6))

ep.plot_bands(lidar_dem_im, 
              ax=ax, 
              cmap='gray',
              title="Challenge 1: Lidar Digital Terrain Model (DTM)\n overlayed on top of a hillshade")

ep.plot_bands(lidar_dem_hill, 
              cmap='gray', 
              alpha=0.8, 
              ax=ax, 
              cbar=False)

plt.show()
```

```python
#Challenge 2: Overlay DTM Over DTM Hillshade

fig, ax = plt.subplots(figsize=(10, 6))

ep.plot_bands(lidar_dem_im, 
              ax=ax, 
              cmap='viridis',
              title="Challenge 2: Lidar Digital Terrain Model (DTM)\n overlayed on top of a hillshade")

ep.plot_bands(lidar_dem_hill, 
              cmap='gray', 
              alpha=0.5, 
              ax=ax, 
              cbar=False)

ax.set_axis_off()

plt.show()
```
